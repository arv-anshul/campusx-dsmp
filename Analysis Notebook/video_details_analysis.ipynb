```python
import pandas as pd
from json import load, dump
```

```python
with open('../data/video_details_new.json') as f:
    video_details = load(f)['video_details']
```

```python
df = pd.DataFrame.from_dict(video_details)
print(df.shape)
df.head()
```

### The these cols because they have only one unique value.

```python
df.drop(columns=['description', 'title', 'channelTitle', 'channelId', 'categoryId', 'tags'],
        inplace=True)
```

```python
df.info()
```

```python
# Datetime cols
df['publishedAt'] = df['publishedAt'].astype('datetime64')
```

```python
# Get length of video in minutes from duration column
hour = df['duration'].str.extract(r'(\d)H').fillna(0).astype(int)
minute = df['duration'].str.extract(r'(\d{2})M').fillna(0).astype(int)

df['duration'] = hour.mul(60).add(minute)
```
